# Facial Recognition: Converting the Image Folders to Train/Test Folders

### This program converts the images generated by OpenCV into train, validation, and test data. This is done to design the cnn model that will predict the faces.

```python
# Library necessary to split the folders
import splitfolders
```

```python
'''
This program currently runs based on my saved location of the folder in question.
To have this run successfully on another computer, simply change 'path' to wherever
the folder with all of your images are.
'''
# Path is set to the location of the image folders
path = r'C:/Users/jpasz/Facial Recognition Images'

# Splitfolders then splits the images into 70% Train, 10% Validation, and 20% Test
splitfolders.ratio(path, output='Facial Recognition TVT', seed=42, ratio=(0.7, 0.1, 0.2), group_prefix=None)
```
