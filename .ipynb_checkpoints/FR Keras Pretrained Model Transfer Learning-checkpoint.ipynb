# Facial Recognition: VGG16 Model with Transfer Learning

### This program is for creating the facial recognition model using the dataset gathered from OpenCV FR Dataset Generation. Essentially, after the user's face data is collected, then a sequential model is created using the folders of all the user's. 

### In this program, transfer learning is going to be used for creating the sequential model. First, the program is going to load the vgg16 model from keras applications. Then the flatten and fully connected layers are added. The final layer has an output equal to the number of recorded users with softmax activation. Once this is done, the model is saved, and a demonstration is provided showing that the model can successfully differentiate between the different user faces.

### Libraries Used:

```python
# Libraries used for OpenCv and path creation/reading
import cv2
import os
from os import listdir

# Keras libraries necessary for the model creation
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential, Model
from keras.layers import Input, Dense, Activation
from keras.layers import Flatten, Dropout
from keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D

# Keras library for VGG16 model
from keras.applications.vgg16 import VGG16

# Libraries for preprocessing the data
from PIL import Image
from keras.preprocessing import image
from keras.preprocessing.image import load_img, img_to_array, save_img
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.imagenet_utils import preprocess_input
from tensorflow.keras.utils import to_categorical
from glob import glob

# Importing numpy to work with arrays, matplotlib for visuals
# and pandas for dataframe.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Loading the pre-trained VGG16 Model from Keras Applications:

```python
# The image shape is simply the dimensions that the images from the OpenCV FR program were saved as.
image_shapes = (224, 224, 3)

# The vgg16 model is loaded from keras applications. The input shape is equal to the image shapes,
# and the imagenet weights are applied. These layers are set to non-trainable for transfer learning
# purposes.
vgg_keras = VGG16(include_top=False, weights='imagenet', input_shape=image_shapes)
vgg_keras.trainable = False
vgg_keras.summary()
```

### Transfer Learning Model Creation:

```python
# The variable num_classes is created. It is equal to the number of users created in the 
# OpenCV FR program.

'''
Adjust path to where your face images are without before using the split-folders program.
Or, to one of the other folders where all of the user image folders are.
'''
num_classes = len(next(os.walk(r'C:\Users\jpasz\Facial Recognition Images'))[1])
print(num_classes)
```

```python
# A sequential model is created here for transfer learning. The layers from vgg16 in keras
# are loaded, then the fully connected layers are added. First, a flatten layer is added to 
# reshape the data, the two Dense layers and Dropout layers are applied. After this,
# a final fully connected output layer is created with the output size of num_classes and
# softmax activation.
frv_model = Sequential()
frv_model.add(vgg_keras)
frv_model.add(Flatten())
frv_model.add(Dense(128, activation='relu'))
frv_model.add(Dropout(0.5))
frv_model.add(Dense(64, activation='relu'))
frv_model.add(Dropout(0.5))
frv_model.add(Dense(num_classes, activation='softmax'))

# The transfer learning model is the compiled with categorical crossentropy loss, adam optimizer,
# and accuracy metrics.
frv_model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
frv_model.summary()
```

```python
'''
For this to work properly, you will have to adjust the path for your images to train, and test 
respectively. To do this, take your image dataset collected from OpenCV FR Dataset Generation,
then run the FR Image Folders to Train/Test program on the github. Set the train and test
paths to the train folder created by the Train/test program. See below...

Also, this program and the split folders program will likely be combined once universal directories
are created, so that these paths do not have to be changed from person to person.
'''

# The train data is created by using a ImageDataGenerator from Keras preprocessing to create more
# data for train/test. The data will be normalized, and have a chance to be flipped horizontally,
# as well as a shear angle of 0.22 that can be applied. The train data is then formed by reading all
# the images from the train directory.
temp_train = ImageDataGenerator(rescale=1.0/255.0, horizontal_flip=True, rotation_range=30,
                                width_shift_range=0.1, height_shift_range = 0.1, shear_range=0.22)
train_data = temp_train.flow_from_directory(r'C:\Users\jpasz\Facial Recognition TVT\train',
                                           target_size=(224, 224), batch_size=16,
                                           class_mode='categorical')
# The test data is created by using a ImageDataGenerator from Keras preprocessing to normalize the
# data for train/test. The test data is then formed by reading all the images from the test directory.
temp_test = ImageDataGenerator(rescale=1.0/255.0)
test_data = temp_test.flow_from_directory(r'C:\Users\jpasz\Facial Recognition TVT\test',
                                         target_size=(224, 224), batch_size=16,
                                         class_mode='categorical')
```

```python
# ReduceLROnPlateau is used to reduce the learning rate when the model is deemed to 
# not be improving anymore. It will monitor validation loss, have a factor of 0.3
# (factor is applied to the rate learning is reduced), and patience will be 4
# (patience is how many iterations without improvement before learning rate is reduced).
reduce_LRoP = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                                                         patience=1, min_lr=0.001)
# EarlyStopping is used to stop the training after a set amount of iterations without
# improvement. <- In this case, the patience is 7. It will monitor validation loss,
# and run in the min mode.
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=2, verbose=1, 
                                           mode='min', baseline=None, restore_best_weights=True)
# The functional model is fitted with the train data, batch size, epochs,
# a validation of the test data, shuffle, and the callbacks from above.
frv_model.fit(train_data, batch_size=32, epochs=6,
             validation_data=test_data, shuffle=True,
             callbacks=[reduce_LRoP, early_stop])
```

```python
# The model is saved as an h5 file to be used for future predictions.
frv_model.save('frm_model' + '.h5')
```

### Model Predictions:

```python
# Now to test the model's accuracy

# The transfer learning model is loaded to demonstrate reloading the model
from keras.models import load_model
tl_model = load_model('frm_model.h5')

# The model is evaluated with the test data.
[test_loss, test_acc]= tl_model.evaluate(test_data)
print("\nEvaluation result on Test Data : Loss = {}, accuracy = {}".format(test_loss, test_acc))
```

```python
# The model is also evaluated with the train data. This was the see if anything changed now that
# the model was reloaded.
[test_loss, test_acc]= tl_model.evaluate(train_data)
print("\nEvaluation result on Test Data : Loss = {}, accuracy = {}".format(test_loss, test_acc))
```

```python
# This was done to show a sample image output from the train data.

# The train data was broken up into images and labels here.
x_sample, x_labels = next(train_data)

# The model makes predictions about all of the images within the train data.
# The first image was outputed below with it's real label, as well as the predicted label.
pred = tl_model.predict(x_sample)
plt.imshow(x_sample[0])
plt.title('The real label is: {}\nThe predicted label is: {}'.format(authorized_names[x_labels[0].argmax()],
                                                                             authorized_names[pred[0].argmax()]))
plt.show()
```

```python
# This for loop simply does this concept from above once again, but with the first 
# 9 images from the train data.
fig, ax = plt.subplots(3, 3, figsize=(14, 14))
pred = tl_model.predict(x_sample)
for i, ax in enumerate(ax.flat):
    ax.imshow(x_sample[i])
    ax.set(title = 'The real label is: {}\nThe predicted label is: {}'.format(authorized_names[x_labels[i].argmax()],
                                                                             authorized_names[pred[i].argmax()]))
```
